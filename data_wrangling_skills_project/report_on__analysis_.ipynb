## Introduction
> After preparation of data gathered through the wrangling process, it was now necessary to conduct an analysis into the information that could be extracted from thecleaned master data. The for the analysis, there were four focus questions.
> > 1. What is the most popular dog name?
> > 2. What is the most popular dog stage?
> > 3. What is the most accurate prediction algorithm and what might influence the algorithms confidence.
> > 4. What sources do the algorithms find most confidence when predicting.

## Analysis and results

>An analysis was done on the focus questions and each was addressed. The results were as follows:
> * The most popular dog name was investigated using groupby and finding the mean of the favourite counts for each dog name. The data shows the most popular dog name is `Stephan` followed by `Jamesy`. 
> * The most popular dog stage was investigated using groupby and finding mean of the favourite counts for each dog name. The data shows dogs on category `doggo, puppo` are the most popular.
> * The most accurate algorithm was investigated by grouping the data by the most confident algorithm and analysing the confidence means. The data shows the most confident algorithm is algorithm 1
> * Most predictions with stronger confidence ratings come from Twitter for Iphone. This may indicate that Twitter for Iphone users may be getting better quality images for neural nertwork prediction algorithm.

```python
# 1. Most popular dog name
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('twitter_archive_master.csv')
plt.figure(figsize = (17,6))
df.groupby('dog_name').mean()['favorite_count'].sort_values(ascending=False).head(5).plot.bar()
plt.title("The most popular dog name",fontsize = 14)
plt.xlabel("Dog names",fontsize = 16);
plt.ylabel("Favourite count",fontsize = 16);
```

```python
#let's extract the sample image of most popular dog
from IPython.display import Image
popular_dog_stage_img = df[df['dog_stage'] == 'doggo, pupper'].sort_values(by = 'favorite_count',ascending = False)['image_link']
# extract just the image url (jpg_url), breed and favorite_count columns
#popular_dog_stage_img = popular_dog_stage_img[['dog_stage','favorite_count','image_link']][0:5]
# view
popular_dog_stage_img.loc[1134]
#get the most popular by favorite count and view
popular_dog_stage_img = popular_dog_stage_img.loc[1134]
popular_dog_stage_img
# Doggo,puppo
Image(popular_dog_stage_img)
```
